```python
data_columns = train_data.columns.to_list()
# print(data_columns)
features = []
for column in data_columns:
    # print(train_data[column].dtype)
    if train_data[column].dtype != 'object':
        features.append(column)
features
```

# MODEL VALIDATION

## Calculate Mean Absolute Error

## Train Test Split

```python
from sklearn.model_selection import train_test_split

# Split data into training and validation data, for both features and target.
# The split is based on a random number generator. Supplying a numeric value
# to the 'random_state' argument guarantees we get the same split everytime
# we run this script.

train_house_features, val_house_features, train_target, val_target = train_test_split(house_features, target, random_state=1)

# Define model
train_model = DecisionTreeRegressor()

# Fit model
train_model.fit(train_house_features, train_target)
```

## Make Predictions With Validation Data

```python
# get predicted prices on validation data
val_predictions = train_model.predict(val_house_features)

# Print the top few validation predictions
print(val_predictions[:5])

# Print the top few actual prices from validation data
print(val_target[:5])
```

## Calculate Mea in Validation Data

```python
val_mae = mean_absolute_error(val_target, val_predictions)

print("The mean absolute error for the out-of-sample data:", val_mae)
```

# OVERFITTING & UNDERFITTING

## Compare Different Tree Size

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    print("MAE:",mae)
    return mae
```

```python
mae_dict = {}
for max_leaf_nodes in range(2,5001):
    my_mae = get_mae(max_leaf_nodes, train_house_features, val_house_features, train_target, val_target)
    mae_dict[max_leaf_nodes] = my_mae
    # print("Max leaf nodes: %d \t\t Mean Absolute Error: %d" % (max_leaf_nodes, my_mae))

best_tree_size = min(mae_dict, key=mae_dict.get)
print("The best Max Leaf Nodes is",best_tree_size,"leaves")
```

## Fit Model Using All Data

```python
# Fill in argument to make optimal size & document
final_model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state=0)

# Fit the final model
final_model.fit(house_features, target)
print(final_model)
```

# BUILDING MODEL (RANDOM FOREST)
